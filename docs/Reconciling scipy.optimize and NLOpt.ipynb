```python
import matplotlib.pyplot as plt
import numpy as np
import nlopt

from scipy.optimize import minimize, rosen, rosen_der
from mpl_toolkits.mplot3d import Axes3D
from matplotlib.colors import LogNorm

from pyalacarte.utils import unvectorize_args
```

```python
rosen_ = unvectorize_args(rosen)
```

```python
y, x = np.mgrid[-1:3.1:0.1, -2:2.2:0.1]
```

```python
z = rosen_(x, y)
```

```python
fig = plt.figure()
ax = plt.axes(projection='3d', azim=-128, elev=43)
ax.plot_surface(x, y, z, norm=LogNorm(), rstride=1, cstride=1, cmap=plt.cm.jet, alpha=0.8, edgecolor='none')

ax.set_xlim([-2, 2.0])                                                       
ax.set_ylim([-1, 3.0])                                                       
ax.set_zlim([0, 2500]) 

ax.set_xlabel('x')
ax.set_ylabel('y')
```
